```python
c2=conn2.cursor()
cursor=c2.execute("SELECT * FROM poet order by id asc limit 2000")
for row in cursor:
    data.append(row)
```

```python
train_data=[]
train_target=[]
for i in range(0,4000):
    train_data.append(data[i][3])
    train_target.append(data[i][4])
print(train_data)
print(train_target)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train_data, train_target,test_size=0.4,random_state=0) #目前还无分类
```

import jieba
import numpy as np
def fenci(train_data):
    words_df = train_data.apply(lambda x:' '.join(jieba.cut(x)))
    return words_df

```python
infile = open("C:\\Users\\洋子\\Documents\\作业\\大三上\\文本理解与数据挖掘\\第一次作业\\stopwords_cn.txt",encoding='utf-8')
stopwords_lst = infile.readlines()
stopwords = [x.strip() for x in stopwords_lst]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
vector = cv.fit_transform(x_train)
vector.todense()
```

```python
cv.vocabulary_
```

```python
vectorizer = CountVectorizer(stop_words=stopwords, max_features=5000)
vectorizer.fit(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
#模型训练
classifier.fit(vectorizer.transform(x_train), y_train)
#使用训练好的模型进行预测
classifier.score(vectorizer.transform(x_test), y_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#使用tf-idf把文本转为向量
tv = TfidfVectorizer(stop_words=stopwords, max_features=5000, lowercase = False)
tv.fit(x_train)
#模型训练
classifier.fit(tv.transform(x_train), y_train)
#利用训练好的模型测试
classifier.score(tv.transform(x_test), y_test)
```

```python
#转词向量
tv_2gram = TfidfVectorizer(stop_words=stopwords, max_features=5000, ngram_range=(1,2),lowercase = False)
tv_2gram.fit(x_train)
#训练模型
clf_2gram = MultinomialNB()
clf_2gram.fit(tv_2gram.transform(x_train), y_train)
#预测
clf_2gram.score(tv_2gram.transform(x_test), y_test)
```
